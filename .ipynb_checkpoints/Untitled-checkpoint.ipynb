```python
from datetime import datetime

now = datetime.utcnow().strftime("%Y%m%d%H%M%S")
root_logdir = "tf_logs"
logdir = "{}/run-{}/".format(root_logdir, now)
```

```python
import tensorflow as tf
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
m, n = housing['data'].shape
data = np.concatenate((np.ones((m, 1)), housing['data']), axis = 1)
```

```python
X = tf.constant(data, dtype = tf.float32)
y = tf.constant(housing['target'].reshape(-1, 1), dtype = tf.float32)

theta = tf.Variable(tf.random_uniform([n + 1, 1], -1.0, 1.0))
y_pred = tf.matmul(X, theta)
error = y_pred - y
mse = tf.reduce_mean(tf.square(error)) 
optimizer =tf.train.AdamOptimizer()
training_op = optimizer.minimize(mse)

mse_summary = tf.summary.scalar('MSE', mse)
file_writer = tf.summary.FileWriter(logdir, tf.get_default_graph())

init = tf.global_variables_initializer()
```

```python
n_epochs = 100000
with tf.Session() as sess:
    init.run()
    
    for epoch in range(n_epochs):
        if epoch % 100 == 0:
            summary_str = mse_summary.eval()
            file_writer.add_summary(summary_str, epoch)
        if epoch % 10000 == 0:
            print("Epoch", epoch, "MSE=", mse.eval())
        sess.run(training_op)
    best_theta = theta.eval()
    file_writer.close()
```

```python
def dense_layer(input_matrix, n_neurons, activation = None):
    n_cols = int(input_matrix.shape[1])
    stddev = 2 / np.sqrt(n_cols)
    init = tf.truncated_normal((n_cols, n_neurons), stddev=stddev)
    
    W = tf.Variable(init)
    b = tf.Variable(tf.zeros([1, n_neurons], dtype = tf.float32))
    Z = tf.matmul(input_matrix, W) + b
    
    if activation == None:
        return Z
    else:
        return activation(Z)
```

```python
X = tf.placeholder(tf.float32, shape = (None, n_cols))
y = tf.placeholder(tf.int64, shape = (None, 10))

hidden1 = dense_layer(X, 300, activation=tf.nn.relu)
hidden2 = dense_layer(hidden1, 100, activation=tf.nn.relu)
logits = dense_layer(hidden2, 10)

xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y_train, logits=logits)
loss = tf.reduce_mean(xentropy)

optimizer = tf.train.AdamOptimizer()
training_op = optimizer.minimize(loss)

init = tf.global_variables_initializer()
```

```python
from tensorflow.keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
from keras.utils import to_categorical
X_train = X_train / 255
y_train.shape
```

```python
m, seq_len, n = X_train.shape
```

```python
n_neurons = 150

from keras.models import Model
from keras.utils import plot_model
from keras.layers import Input, SimpleRNN, Dense

# Define an input sequence and process it.
inputs = Input(shape=(seq_len, n))
h1 = SimpleRNN(n_neurons)(inputs)
prob = Dense(10, activation = 'softmax')(h1)

model = Model(inputs=inputs, outputs=prob)
# summarize layers
print(model.summary())
# plot graph
plot_model(model, to_file='recurrent_neural_network.png')
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
```

```python
model.fit(X_train, y_train, validation_split=0.2,batch_size = 2 ** 13, epochs=5) 
#pred = model.predict(test_x)
```

```python
n_neurons = 100
n_steps = 100

from keras.models import Model
from keras.utils import plot_model
from keras.layers import Input, SimpleRNN, Dense, TimeDistributed, Reshape

# Define an input sequence and process it.
inputs = Input(shape=(n_steps, 1))
h1 = SimpleRNN(n_neurons, return_sequences = True)(inputs)
pred = TimeDistributed(Dense(1), input_shape = (n_steps, n_neurons))(h1)

model = Model(inputs=inputs, outputs=pred)
# summarize layers
print(model.summary())
# plot graph
plot_model(model, to_file='recurrent_neural_network.png')
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
X_train = []
Y_train = []
for k in range(1000 - n_steps):
    X_train.append(time[k : k + n_steps])
    Y_train.append(value[k + 1 : k + 1 + n_steps])

X_train = np.array(X_train)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
Y_train = np.array(Y_train)
Y_train = Y_train.reshape(Y_train.shape[0], Y_train.shape[1], 1)
```

```python
model.fit(X_train, Y_train, batch_size = 512, epochs=1000)
```

```python
pred_val = []

init_time = time[:n_steps]
#init_val = value[:n_steps]
init_val = np.exp(init_time) 

for i in range(n_steps):
    pred_val.append(init_val[i])
for i in range(n_steps, len(time)):
    data = np.array(pred_val[-n_steps : ]).reshape((1, n_steps, 1))
    next_val = model.predict(data)[0, -1, 0]
    pred_val.append(next_val)
```
